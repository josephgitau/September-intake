```python
# Import the necessary packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier # KNeighborsRegressor for regression
```

```python
# import the iris dataset from seaborn
iris = sns.load_dataset('iris')
iris.head()
```

```python
# check unique species
iris['species'].unique()
```

```python
# drop rows with class label 'setosa'
# iris = iris[iris['species'] != 'versicolor']
```

```python
# split the data into features and target
X = iris.drop('species', axis=1)
y = iris['species']
```

```python
y_map = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
y_changed = y.map(y_map)

# Plot the dataset
plt.scatter(X['sepal_length'], X['petal_length'], c=y_changed, cmap='viridis')
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.title('Iris Dataset')
plt.show()
```

```python
# Create a KNeighborsClassifier model
knn = KNeighborsClassifier(n_neighbors=8) # n_neighbors is the hyperparameter (we can change this value and it will affect the model accuracy)

# Fit the model to the data
knn.fit(X, y)

# model accuracy
print('Model Accuracy:', knn.score(X, y))
```

```python
# fit different knn models and we can pick the n value with the highest accuracy
n = np.arange(3, 15)
accuracy = []

for i in n:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X, y)
    print(f'Accuracy for n={i}:', knn.score(X, y))
    accuracy.append(knn.score(X, y))

# print the n with the highest accuracy
print('n with the highest accuracy:', n[accuracy.index(max(accuracy))])
```
