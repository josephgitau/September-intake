## Simple Linear Regression

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generating data
df = pd.DataFrame({"Number_of_customers":np.random.randint(1,101,100),
                   "Sales":np.random.randint(100,1001,100)})

df.head()
```

## Problem Statement

In this codebook we will try predict the Sales based on the number of customers we have.

```python
# create a line plot of the data
plt.plot(df["Sales"])
plt.xlabel("Observations")
plt.ylabel("Sales amount")
plt.title("Sales trend")
```

```python
# create a histogram of the sales data
plt.hist(df["Sales"], bins=15);
```

## Modelling

```python
# create our target and feature variables
X = df[["Number_of_customers"]]
y = df["Sales"]

# import our model
from sklearn.linear_model import LinearRegression

# instantiate our model
lr = LinearRegression()

# fit our model
lr.fit(X, y)

# make predictions
y_pred = lr.predict(X)

# score your model
print("Model score: ", lr.score(X, y))
```

```python
# extract the model coefficients and intercept
print("Model coefficients: ", lr.coef_)
print("Model intercept: ", lr.intercept_)
```
