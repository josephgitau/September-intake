```python
# import libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
# Load the dataset
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
```

```python
# Create a dataframe
data = pd.DataFrame(X, columns=cancer.feature_names)
data['target'] = y

# Inspect the data
data.head()
```

```python
# Summary statistics
data.describe()
```

```python
# Check for missing values
data.isnull().sum()
```

```python
# Visualize the data
sns.countplot(data, x='target')
plt.show()
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier # Base model
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Prepare the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# fit a decision tree model
model = DecisionTreeClassifier(random_state=42)

# fit the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
```

```python
# visualize the decision tree
from sklearn.tree import plot_tree
plt.figure(figsize=(20,10))
plot_tree(model, feature_names=cancer.feature_names, filled=True)
plt.show()
```

```python
# Fit Bagging

# specify our base model
base_model = DecisionTreeClassifier(random_state=42)

# instantiate the bagging model
bagging = BaggingClassifier(base_estimator=base_model, n_estimators=100, random_state=42)

# Fit the model
bagging.fit(X_train, y_train)

# Make predictions
y_pred = bagging.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression

# Create base models
logistic_regression = LogisticRegression(max_iter=10000) 

# Create the bagging model
bagging = BaggingClassifier(logistic_regression, n_estimators=100, random_state=42)

# Fit the model
bagging.fit(X_train, y_train)

# Make predictions
y_pred = bagging.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```
